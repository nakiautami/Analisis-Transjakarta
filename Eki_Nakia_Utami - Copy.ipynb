## 3.4. Missing Value

Penanganan Missing Value:

Ada dua cara yang digunakan untuk menangani missing value:
* Mengisi missing value dengan nilai yang sesuai.
* Menghapus baris yang missing value-nya tidak bisa diisi.

Pengisian missing value dilakukan dengan memanfaatkan kolom lain yang memiliki hubungan melalui metode mapping. Proses pengisian mengikuti aturan berikut:
* Jika kolom acuan (benchmark) tidak kosong, maka missing value diisi menggunakan mapping dari kolom tersebut.
* Jika kolom acuan kosong, missing value diisi menggunakan mapping dari kolom lain, atau baris tersebut akan dihapus (drop).

## 3.4.1 Missing Values corridorID menggunakan mapping corridorName

* Kolom `corridorID` memiliki missing values sebesar 3.3%.`corridorID` yang kosong akan diisi mengunakana data `corridorName` karena 2 kolom ini saling mempresentasikan satu sama lain, artinya setiap satu `corridorID` mewakili satu `corridorName` yang sama.

## 4.5. Analisis Origin Destination (OD) pada Transjakarta

Sebelum analsis lebih lanjut kita perlu mengetahui istilah-istilah yang akan banyak kita gunakan.

Berikut adalah perbedaan antara halte, koridor, dan rute dalam konteks sistem transportasi TransJakarta:

1. **Halte
Definisi**: Halte adalah titik pemberhentian bus, tempat penumpang naik atau turun dari bus. Halte bisa berada di sepanjang jalan atau di tempat tertentu di dalam koridor.
Fungsi: Halte berfungsi sebagai tempat transit bagi penumpang. Penumpang akan menunggu di halte untuk naik atau turun dari bus.
Contoh: Halte Blok M, Halte Kota, Halte Dukuh Atas, dll.
2. **Koridor
Definis**i: Koridor adalah jalur atau trayek bus yang menghubungkan beberapa halte di sepanjang rute tertentu. Koridor dapat melibatkan beberapa halte dan sering kali diatur dalam jalur busway yang terpisah dari jalan utama.
Fungsi: Koridor mencakup keseluruhan jalur perjalanan yang menghubungkan berbagai titik pemberhentian (halte) dan menyediakan jalur tetap bagi bus TransJakarta. Koridor memberikan gambaran lebih besar tentang jalur perjalanan antar area yang lebih luas.
Contoh: Koridor 1 yang menghubungkan Blok M dengan Kota atau Koridor 3 yang menghubungkan Kalideres dengan Cawang.
3. **Rute
Definisi**: Rute adalah jalur spesifik yang dilalui oleh bus, yang bisa mencakup beberapa koridor. Rute mencakup perjalanan dari satu titik asal ke tujuan tertentu dan bisa melibatkan beberapa koridor atau segmen perjalanan.
Fungsi: Rute mendefinisikan perjalanan bus tertentu dari titik awal hingga tujuan akhir, termasuk semua halte dan koridor yang dilalui dalam perjalanan tersebut.
Contoh: Rute yang menghubungkan Cawang ke Kota bisa melibatkan Koridor 3 dan Koridor 1.


Perbedaan Utama:
* Halte adalah titik pemberhentian di sepanjang perjalanan.
* Koridor adalah jalur panjang yang menghubungkan beberapa halte.
* Rute adalah perjalanan spesifik bus yang mencakup beberapa koridor dan halte dari asal ke tujuan.
Dengan kata lain, **halte adalah tempat berhenti, koridor adalah jalur utama yang menghubungkan beberapa halte, dan rute adalah perjalanan bus yang mencakup beberapa koridor dan halte.**

## 4.5.3.4 Rasio Efektivitas Koridor

Rasio efektivitas koridor mengukur persentase penumpang yang menyelesaikan rute dengan membandingkan jumlah penumpang yang melakukan tap-in pada halte awal (stopStartSeq=1) dengan penumpang yang melakukan tap-out pada halte akhir (stopEndSeq=1).

```python
# Filter data untuk tap-in di halte awal (stopStartSeq=1)
tap_in_start = cleaned_tj[cleaned_tj["stopStartSeq"] == 1].shape[0]

# Filter data untuk tap-out di halte akhir (stopEndSeq=1)
tap_out_end = cleaned_tj[cleaned_tj["stopEndSeq"] == 1].shape[0]
print(f"Jumlah tap-in di halte awal: {tap_in_start}")
print(f"Jumlah tap-out di halte akhir: {tap_out_end}")
```

```python
# Hitung rasio efektivitas
if tap_in_start > 0:
    rasio_efektivitas = (tap_out_end / tap_in_start) * 100
else:
    rasio_efektivitas = 0  # Untuk menghindari pembagian nol
print(f"Rasio efektivitas koridor: {rasio_efektivitas:.2f}%")
```

```python
categories = ['Tap-In (Start)', 'Tap-Out (End)']
values = [tap_in_start, tap_out_end]

# Membuat grafik
plt.figure(figsize=(8, 6))  # Perbesar ukuran grafik agar lebih leluasa
bars = plt.bar(categories, values, color=['cornflowerblue', 'salmon'], edgecolor='black')

# Menambahkan label nilai pada batang
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 2,
             f'{int(bar.get_height())}', ha='center', color='black', fontsize=12)

# Menambahkan keterangan rasio dekat dengan judul
plt.title('Efektivitas Koridor: Tap-In vs Tap-Out\n', fontsize=15, pad=10)
plt.text(0.5, max(values) + 15, f'Rasio Efektivitas: {rasio_efektivitas:.2f}%', 
         ha='center', fontsize=13, fontweight='bold', color='darkgreen')

# Menyesuaikan jarak antara keterangan rasio dan grafik
plt.ylim(0, max(values) + 10)  #

# Pengaturan label sumbu dan grid
plt.xlabel('Kategori', fontsize=12)
plt.ylabel('Jumlah Penumpang', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Tampilkan grafik
plt.tight_layout()
plt.show()
```

     penumpang tidak menyelesaikan perjalanan mereka hingga halte akhir.

## 4.5.5 Analisis Halte

Tujuan: Menemukan halte yang paling banyak dikunjungi  dalam setiap koridor berdasarkan jumlah total tapIn (masuk) dan tapOut (keluar).

```python
# Menggabungkan tapInStopsName dan tapOutStopsName
all_stops = pd.concat([cleaned_tj["tapInStopsName"], cleaned_tj["tapOutStopsName"]])

# Menghitung total aktivitas per halte
stop_activity = all_stops.value_counts().reset_index()
stop_activity.columns = ["Halte", "Total Transaksi"]

# Menghitung jumlah unik tapIn dan tapOut untuk setiap halte
unique_tapIn = cleaned_tj["tapInStopsName"].value_counts().reset_index()
unique_tapIn.columns = ["Halte", "Total TapIn"]

unique_tapOut = cleaned_tj["tapOutStopsName"].value_counts().reset_index()
unique_tapOut.columns = ["Halte", "Total TapOut"]

# Menggabungkan data unik tapIn dan tapOut
corridor_active = pd.merge(unique_tapIn, unique_tapOut, on="Halte", how="outer").fillna(0)

# Menambahkan aktivitas total untuk hasil unik
corridor_active["Total Aktivitas"] = corridor_active["Total TapIn"] + corridor_active["Total TapOut"]
```

```python
# Mengurutkan berdasarkan Total Aktivitas
sort_Aktivitas = corridor_active.sort_values(by=["Total Aktivitas"], ascending=False)
sort_Aktivitas
```

```python
# Menghitung total aktivitas per koridor
corridor_totals = corridor_active.groupby('Halte')['Total Aktivitas'].sum().reset_index()
corridor_totals_sorted = corridor_totals.sort_values(by='Total Aktivitas', ascending=False)
top_10_corridors = corridor_totals_sorted.head(10)

# Plotting
plt.figure(figsize=(12, 8))

# Custom color palette for bars
colors = cm.tab10(np.linspace(0, 1, len(top_10_corridors['Halte'])))
bars = plt.bar(top_10_corridors['Halte'], top_10_corridors['Total Aktivitas'], color=colors)

# Adding labels for each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval), va='bottom', ha='center')

plt.xlabel('Halte Name')
plt.ylabel('Total Transactions')
plt.title('Top 10 Halte by Total Transactions', fontsize=20, fontweight='bold')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

## 7.1. Analisis Pengaruh Ketidakseimbangan Kapasitas Rute terhadap Efisiensi Operasional 


Evaluasi ketidakseimbangan kapasitas rute Transjakarta dilakukan dengan menganalisis data jumlah penumpang, kapasitas armada, dan frekuensi layanan pada setiap rute. Tujuan evaluasi ini adalah:
* Mengidentifikasi rute overcapacity, yaitu rute di mana 
  jumlah penumpang melebihi kapasitas armada, yang menyebabkan kepadatan, waktu tunggu yang lama, dan penurunan kenyamanan penumpang.
* Mengidentifikasi rute undercapacity, yaitu rute dengan 
  kapasitas armada yang tidak optimal, sehingga ada kapasitas yang tidak terpakai dan meningkatkan biaya operasional tanpa manfaat yang seimbang.
* Mengevaluasi dampaknya terhadap efisiensi operasional, termasuk waktu tunggu penumpang, kenyamanan, dan biaya operasional.

Data yang dibutuhkan dalam analisis:
1.  Data Rute dan Armada
    * Daftar rute (kode, nama, panjang rute dalam km).
    * Jumlah armada per rute.
    * Kapasitas maksimum per bus (jumlah penumpang). (data ini tidak tersedia di data set)
    * Frekuensi keberangkatan bus per rute per jam.
2. Data Penumpang
    * Volume penumpang per rute per hari/jam.
    * Volume penumpang per halte per rute (untuk analisis 
      waktu tunggu).
3. Indikator Operasional
    * Waktu tunggu rata-rata per halte/rute.
    * Waktu tempuh rata-rata per rute.
    * Tingkat keterlambatan bus.
    
Catatan: Data kapasitas maksimum per bus penting untuk menghitung kapasitas total armada.

Analisis ini dilakukan berdasarkan indikator **Load Factor**, yaitu tingkat pemanfaatan kapasitas armada pada setiap rute.
load factor = Jumlah Penumpang/ (jumlah armada X kapasitas Bus)

Nilai load factor ini yang akan menentukan **overcapacity atau undercapacity**.

Berdasarkan hasil analisis nya dapat dilakukan simulasi strategi yaitu:
* Penyesuaian jummlah armada dapat dilakukan dengna 
  redistribusi armada dari rute undercapacity ke rute overcapacity.
* Tambahkan frekuensi keberangkatan bus pada rute 
  overcapacity untuk mengurangi waktu tunggu penumpang.


Analisis ini dapat memberikan keuntungan strategis dan finansial bagi Transjakarta melalui langkah-langkah berikut:
1. Optimasi Alokasi Sumber Daya
   * Mengurangi pemborosan dengan redistribusi armada  dari rute undercapacity ke overcapacity.
   * Meningkatkan efisiensi operasional melalui pengoperasian armada pada kapasitas optimal.

2. Peningkatan Pengalaman Penumpang
   * Mengurangi waktu tunggu dan kepadatan, meningkatkan kenyamanan perjalanan.
   * Mendorong loyalitas dan retensi penumpang untuk meningkatkan pendapatan.

3. Penurunan Biaya Operasional
   * Menghindari pengeluaran yang tidak perlu seperti bahan bakar, perawatan, dan gaji pengemudi di rute undercapacity.

4. Peningkatan Pendapatan
   * Memaksimalkan kapasitas di rute padat untuk melayani lebih banyak penumpang.
   * Mengidentifikasi potensi restrukturisasi rute untuk meningkatkan profitabilitas.

## 7.2. Analisis Waktu Tunggu 


**Mengidentifikasi penyebab utama waktu tunggu yang lama di beberapa halte Transjakarta, khususnya pada rute sibuk.**

**Tujuan Analisis**
Analisis ini bertujuan untuk mengidentifikasi dan mengatasi faktor-faktor yang mempengaruhi waktu tunggu penumpang di Transjakarta, seperti frekuensi bus, masalah teknis, ketidaksesuaian jadwal, dan ketidakseimbangan kapasitas armada dengan kebutuhan penumpang, terutama pada jam sibuk. Dengan membandingkan waktu tap-in penumpang (tapInHour) dengan waktu kedatangan bus (arrivalTime), Transjakarta dapat merencanakan perbaikan strategi operasional.

**Data yang Digunakan**
1. Waktu Tap-In Penumpang (tapInHour): Menunjukkan waktu saat penumpang tap-in untuk memulai perjalanan.
2. Waktu Kedatangan Bus (arrivalTime): Waktu kedatangan bus di stasiun untuk rute tertentu.
3. Identifikasi Stasiun dan Rute: Data rute dan stasiun yang dilalui bus.
4. Data Frekuensi Bus: Frekuensi bus yang melayani rute tertentu.
5. Data Kapasitas Armada dan Penumpang: Jumlah bus dan kapasitas armada pada rute tertentu.

**Data Kapsitas Armada dan jadwal kedatangan bus disini cukup penting untuk analisis**

**Metode Analisis**
1. Perbandingan Waktu Tap-In dan Kedatangan: Menghitung durasi waktu tunggu dengan membandingkan waktu tap-in dan kedatangan bus.
2. Pencarian Pola Keterlambatan: Mencari pola keterlambatan atau ketidaksesuaian jadwal.
3. Identifikasi Jam Sibuk: Menganalisis waktu tunggu pada jam sibuk untuk merencanakan penambahan armada atau perbaikan jadwal.

**Hasil Analisis**
1. Penurunan Waktu Tunggu: Mengidentifikasi dan memperbaiki ketidaksesuaian jadwal untuk mengurangi waktu tunggu.
2. Peningkatan Kapasitas Armada: Menambah bus jika kapasitas armada tidak mencukupi.
3. Perbaikan Jadwal: Mengoptimalkan jadwal bus agar penumpang tidak menunggu lama.
4. Optimalisasi Rute: Menyesuaikan rute yang memiliki waktu tunggu lebih lama.

**Keuntungan Finansial dan Strategi:** 

Pengurangan Biaya Operasional: Dengan mengoptimalkan frekuensi bus dan penjadwalan, Transjakarta dapat mengurangi biaya operasional yang disebabkan oleh keterlambatan atau kelebihan kapasitas yang tidak efisien.

Peningkatan Kepuasan Pelanggan: Penurunan waktu tunggu dan peningkatan kenyamanan pelanggan dapat meningkatkan kepuasan penumpang, yang pada gilirannya dapat meningkatkan jumlah penumpang yang menggunakan layanan Transjakarta, meningkatkan pendapatan dari tarif.

Efisiensi Sumber Daya: Dengan analisis yang baik mengenai waktu tunggu dan kapasitas armada, Transjakarta dapat lebih efisien dalam mengalokasikan sumber daya seperti bus dan sopir untuk rute yang membutuhkan.

mplementasi Sistem Prediksi Waktu Kedatangan: Transjakarta dapat mengembangkan sistem yang memprediksi waktu kedatangan bus dengan lebih akurat, menggunakan data real-time untuk menginformasikan penumpang mengenai waktu kedatangan yang lebih tepat.
Peningkatan Integrasi Data: Memperkuat integrasi data antara sistem tap-in dan jadwal kedatangan bus untuk memperoleh data yang lebih akurat dan dapat dianalisis dengan lebih baik.
